# 01 — Data Cleaning (All Stations)

This notebook performs data cleaning on all raw water quality datasets
obtained from Google Earth Engine. The same preprocessing steps are applied
consistently to every monitoring station.


Cell 2 — Import Library

```python
import os
import pandas as pd
```

Cell 3 — Define Paths

```python
RAW_DIR = "../data/raw"
CLEAN_DIR = "../data/cleaned"

os.makedirs(CLEAN_DIR, exist_ok=True)
```

Cell 4 — Define Columns to Keep & Rename

```python
COLUMN_MAP = {
    "Date": "date",
    "Secchi_Depth": "secchi",
    "Chlorophyll_a": "chlorophyll_a",
    "TSI": "tsi",
    "Turbidity": "turbidity",
    "Salinity": "salinity",
    "DO": "do",
    "pH": "ph"
}

SELECTED_COLUMNS = list(COLUMN_MAP.keys())
```

Cell 5 — Loop Clean All Files

```python
for file in os.listdir(RAW_DIR):
    if not file.endswith(".csv"):
        continue

    station = file.split("_")[0]
    path = os.path.join(RAW_DIR, file)

    df = pd.read_csv(path)

    # เลือกเฉพาะ column ที่ใช้
    df = df[SELECTED_COLUMNS]

    # rename column
    df = df.rename(columns=COLUMN_MAP)

    # parse date
    df["date"] = pd.to_datetime(df["date"])

    # sort by time
    df = df.sort_values("date")

    # reset index
    df = df.reset_index(drop=True)

    # save
    out_path = os.path.join(CLEAN_DIR, f"{station}_clean.csv")
    df.to_csv(out_path, index=False)

    print(f"✅ Cleaned: {file} → {station}_clean.csv")
```

## Summary

- All raw datasets were cleaned using the same preprocessing pipeline.
- Column names were standardized to snake_case format.
- Temporal information was parsed and sorted consistently.
- Cleaned datasets were saved separately from raw data.
